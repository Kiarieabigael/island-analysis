```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('largest-islands.csv')
df.head()
```

```python
top_10_tropical = (
    df[df["climate"] == "tropics"]
    .sort_values(by="area", ascending=False)
    .head(10)
)

top_10_tropical
```

```python
largest_by_region = df.loc[
    df.groupby("region")["area"].idxmax()
]

largest_by_region
```

```python


df_sorted = df.sort_values(by="rank")

plt.figure()
plt.plot(df_sorted["rank"], df_sorted["area"])
plt.xlabel("Rank (Largest â†’ Smallest)")
plt.ylabel("Area")
plt.title("Island Area vs Rank")
plt.show()
```
